# 3. Creating a regional segmentation of the data

```python
# Regions
northeast = ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']
midwest = ['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']
south = ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']
west = ['Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico', 'Alaska', 'Washington', 'Oregon', 'California', 'Hawaii']
```

```python
# State Regions
df_opc.loc[df_opc['state'].isin(northeast), 'region'] = 'Northeast'
df_opc.loc[df_opc['state'].isin(midwest), 'region'] = 'Midwest'
df_opc.loc[df_opc['state'].isin(south), 'region'] = 'South'
df_opc.loc[df_opc['state'].isin(west), 'region'] = 'West'
```

```python
# checking number in each region
df_opc['region'].value_counts(dropna = False)
```

- Checking the spending habits

```python
# Creating a crosstab
crosstab = pd.crosstab(df_opc['region'], df_opc['spending_flag'], dropna = False)
```

# 4. Create an exclusion flag for low-activity customers

```python
# high activity flag
df_opc.loc[df_opc['max_order'] >= 5, 'activity_flag'] = 'High-activity'
```

```python
# low activity flag
df_opc.loc[df_opc['max_order'] < 5, 'activity_flag'] = 'Low-activity'
```

```python
df_opc.head(15)
```

```python
#verifying
df_opc['activity_flag'].value_counts(dropna = False)
```

```python
# creating df with only active customers
df_high_activity = df_opc.loc[df_opc['activity_flag'] == 'High-activity']
```
